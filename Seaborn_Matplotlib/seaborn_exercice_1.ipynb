# Visualisation d'un nuage de points et exploration des variables

### Importez les librairies

```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### Donnez la liste des variables

```python
house_pricing = pd.read_csv('c:/temp/data/house_pricing.csv');
house_pricing.info()
```

### Cr√©er un nuage de points

```python
sns.scatterplot(x="SalePrice", y="LotArea", data=house_pricing)
plt.show()
```

### Affinez la visualisation

```python
filter = house_pricing["LotArea"] < 20000
filter2 = house_pricing["SalePrice"] < 500000
hp = house_pricing.copy()
hp.where((filter & filter2), inplace = True)
sns.scatterplot(x="SalePrice", y="LotArea", data= hp)
plt.show()
```

### Relation LotFrontage / LotArea

```python
sns.relplot(x="LotFrontage", y="LotArea", data=house_pricing)
plt.show()
```

```python
filter = house_pricing["LotArea"] < 10000
filter2 = house_pricing["LotFrontage"] < 200
house_pricing.where((filter & filter2), inplace = True)
sns.relplot(x="LotFrontage", y="LotArea", data=house_pricing)
```
